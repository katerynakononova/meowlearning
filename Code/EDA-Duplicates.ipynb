# EDA

```python
import pandas as pd

# Load the CSV file
data = pd.read_csv('catfood.csv')

# Show general information about the dataset
print(data.info())
```

```python
# Show the first rows
data.head(20)
```

## Duplicates

```python
# Check for duplicates
duplicates = data.duplicated()
print(f"Number of duplicate rows: {duplicates.sum()}")
```

```python
# Remove duplicates
data = data.drop_duplicates()

# Check the dataset shape after removing duplicates
print(f"Dataset shape after removing duplicates: {data.shape}")
```

```python
data.to_csv('catfood_no_duplicates.csv', index=False)
```
