# EDA

```python
import pandas as pd

# Load the CSV file
data = pd.read_csv('cat_food_data.csv')

# Look at the data
data.head(50)
```

```python
# Check the structure and summary of the data
data.info()
```

```python
# Remove duplicate rows
data = data.drop_duplicates()

# Confirm removal
print(f"Number of rows after removing duplicates: {len(data)}")
```

```python
data.describe()
```

```python
# Count missing values in each column
missing_values = data.isnull().sum()
missing_values[missing_values > 0]  # Display columns with missing values only
```

```python
data = data.dropna()
```

```python
# Check unique values for categorical columns (e.g., 'Brand', 'Flavor')
categorical_columns = data.select_dtypes(include=['object']).columns
for col in categorical_columns:
    print(f"Unique values in {col}:")
    print(data[col].unique())
    print("\n")
```

```python
# Group by 'Brand' and calculate mean, median, and other stats for each numerical column
data.groupby('company').agg(['mean', 'median', 'std'])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot counts of categorical columns
categorical_cols = data.select_dtypes(include=['object']).columns
for col in categorical_cols:
    plt.figure(figsize=(8, 8))
    sns.countplot(y=data[col], order=data[col].value_counts().index)
    plt.title(f'Distribution of {col}')
    plt.xlabel('Count')
    plt.ylabel(col)
    plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filter to include only companies with more than 10 products
company_counts = data['company'].value_counts()
filtered_data = data[data['company'].isin(company_counts[company_counts >= 8].index)]

# Plotting weight vs. price for each company
plt.figure(figsize=(12, 8))
sns.scatterplot(data=filtered_data, x='weight', y='price', hue='company', palette='tab10', alpha=.7)
plt.title('Relationship Between Weight and Price for Companies with More Than 10 Products')
plt.xlabel('Weight')
plt.ylabel('Price')
plt.legend(title='Company', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# Correlation matrix heatmap for numerical columns
numeric_cols = data.select_dtypes(include=['float64', 'int64']).columns
plt.figure(figsize=(10, 8))
sns.heatmap(data[numeric_cols].corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix of Numerical Columns')
plt.show()
```
