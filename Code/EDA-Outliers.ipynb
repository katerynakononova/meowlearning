# EDA

```python
import pandas as pd

# Load the CSV file
data = pd.read_csv('catfood_no_missing.csv')

# Show general information about the dataset
print(data.info())
```

## Outliers

```python
data.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))
sns.boxplot(y=data['price'])
plt.title('Boxplot of Price')
plt.ylabel('Price')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='weight', y='price', hue='company', palette='Set2')
plt.title('Price vs Weight by Brand')
plt.xlabel('Weight')
plt.ylabel('Price')
plt.legend(title='Brand', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```
